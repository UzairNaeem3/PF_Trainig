###  1) Inputs percentage and image, and returns the image by adding white/gray/black border. Border size should be according to the percentage of width and height.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def add_border(img, percentage, color):
    # Get the height and width of the image
    height, width = img.shape[:2]

    # Calculate the size of the border based on the input percentage
    border_size = int(min(height, width) * percentage / 100)

    # Choose the color of the border
    if color == 'white':
        border_color = (255, 255, 255)
    elif color == 'gray':
        border_color = (128, 128, 128)
    elif color == 'black':
        border_color = (0, 0, 0)
    else:
        print(f"Invalid Color: {color}")

    # Create a border around the image using the chosen color and size
    bordered_img = cv2.copyMakeBorder(img, border_size, border_size, border_size, border_size,
                                      borderType=cv2.BORDER_CONSTANT, value=border_color)
    
    
    # Show the  image in new window
    cv2.imshow('Original Image', img)
    cv2.imshow('Bordered Image', bordered_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    return bordered_img

if __name__ == '__main__':
    # Load the input image
    img = cv2.imread(r'C:\Users\uzair\Desktop\Computer Vision\OpenCV_Logo.png')
    color = input("Enter the color: ")

    bordered_img = add_border(img, 10, color)

    # Display the original and bordered image side by side using matplotlib
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
    ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Original Image')
    ax[0].axis('off')
    ax[1].imshow(cv2.cvtColor(bordered_img, cv2.COLOR_BGR2RGB))
    ax[1].set_title('Bordered Image')
    ax[1].axis('off')
    # Add a title to the entire figure
    fig.suptitle('Adding border around the image')
    plt.show()
```

###  2) Inputs image, and resize the image with fix width size 1024 pixels and the heights should be adjust according to the aspect ratio of the image

```python
def resize_image(img):
    # Get the current size of the image
    height, width = img.shape[:2]
    
    # Calculate the aspect ratio of the image
    aspect_ratio = width / height
    
    # Calculate the new height based on the fixed width of 1024 pixels
    new_width = 1024
    new_height = int(new_width / aspect_ratio)
    
    # Resize the image to a width of 1024 pixels while maintaing the aspect ratio
    resized_img = cv2.resize(img, (new_width, new_height))
    
    

    
    return resized_img

if __name__ == '__main__':
    # Load the input image
    img = cv2.imread(r'C:\Users\uzair\Desktop\Computer Vision\OpenCV_Logo.png')

    # Resizing the image
    resized_img = resize_image(img)
    
    
    # Show the  image in new window
    cv2.imshow('Original Image', img)
    cv2.imshow('Resized Image', resized_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
#      # Display the original and resized images side by side using matplotlib
#     fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
#     ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#     ax[0].set_title('Original Image')
#     ax[0].axis('off')
#     ax[1].imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
#     ax[1].set_title('Resized Image')
#     ax[1].axis('off')
#     # Add a title to the entire figure
#     fig.suptitle('Resizing the image')
#     plt.show()
```

### 3) Inputs an image and multiplying factor, and returns the resized image according to multiplying factor.

```python
def resize_image(img, mul_factor):
    height, width = img.shape[:2]
    
    new_width = int(width * mul_factor)
    new_height = int(height * mul_factor)
    
    resized_image = cv2.resize(img, (new_width, new_height))
    
    return resized_image

if __name__ == "__main__":
    
     # Load the input image
    img = cv2.imread(r'C:\Users\uzair\Desktop\Computer Vision\OpenCV_Logo.png')

    # Resizing the image
    resized_img = resize_image(img, 5)
    
    # Show the  image in new window
    cv2.imshow('Original Image', img)
    cv2.imshow('Resized Image', resized_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    
#     # Display the original and resized images side by side using matplotlib
#     fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
#     ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#     ax[0].set_title('Original Image')
#     ax[0].axis('off')
#     ax[1].imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
#     ax[1].set_title('Resized Image')
#     ax[1].axis('off')
#     # Add a title to the entire figure
#     fig.suptitle('Resizing the image')
#     plt.show()
    
```

### Inputs an image and add blur on it

```python
def add_blur(image, kernel):
    blurred_image = cv2.GaussianBlur(image, (kernel, kernel), 0)
    return blurred_image

if __name__ == "__main__":

    # Load the image
    img = cv2.imread(r'C:\Users\uzair\Desktop\Computer Vision\OpenCV_Logo.png')

    # Apply blur with kernel size of 5
    blurred_img = add_blur(img, 9)

    # Show the original and blurred images side by side
    cv2.imshow('Original Image', img)
    cv2.imshow('Blurred Image', blurred_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # Display the original and resized images side by side using matplotlib
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
    ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Original Image')
    ax[0].axis('off')
    ax[1].imshow(cv2.cvtColor(blurred_img, cv2.COLOR_BGR2RGB))
    ax[1].set_title('Blurred Image')
    ax[1].axis('off')
    # Add a title to the entire figure
    fig.suptitle('Adding blur to the image')
    plt.show()
```

### Inputs an image and change brightness of that image

```python
import cv2
import numpy as np

def brightness(image, brightness):
    # Convert the image to the HSV color space
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Split the image into its separate channels
    h, s, v = cv2.split(hsv)
    
    # Apply the brightness adjustment to the V channel
    v = np.clip(v + brightness, 0, 255)
    
    # Merge the channels back together and convert back to the BGR color space
    hsv = cv2.merge((h, s, v))
    brightened_image = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    
    return brightened_image

if __name__ == "__main__":

    # Load the image
    img = cv2.imread(r'C:\Users\uzair\Desktop\Computer Vision\OpenCV_Logo.png')

    # Adjusting the brightness
    brightened_image = brightness(img, 200)

    # Show the original and brightened image side by side
    cv2.imshow('Original Image', img)
    cv2.imshow('Brighted Image', brightened_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
    # Display the original and brightened images side by side using matplotlib
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
    ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax[0].set_title('Original Image')
    ax[0].axis('off')
    ax[1].imshow(cv2.cvtColor(brightened_image, cv2.COLOR_BGR2RGB))
    ax[1].set_title('Brightened Image')
    ax[1].axis('off')
    # Add a title to the entire figure
    fig.suptitle('Adjusting the brightness of the image')
    plt.show()
```

#### Guidelines:

#### * Apply a thresholding method to segment the foreground text from the background. You can use any thresholding technique such as cv2.THRESH_BINARY, cv2.ADAPTIVE_THRESH_MEAN_C, etc.

#### * Preprocess the image using morphological operations such as erosion, dilation, opening, and closing. These operations can be performed using functions such as cv2.erode(), cv2.dilate(), cv2.morphologyEx(), etc. These operations will help in removing noise and filling gaps in the text.

#### * Apply a filter such as cv2.medianBlur() or cv2.bilateralFilter() to smooth the image and further reduce noise.

```python
import cv2

# Load the input image
img_gray = cv2.imread('blur.jpg', cv2.IMREAD_GRAYSCALE)
# image = cv2.imread('blur.jpg')
# image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)


# Apply thresholding to segment the foreground text from the background
thresh = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 5)

# Perform morphological operations to remo]ve noise and fill gaps in the text
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=1)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, kernel, iterations=1)

# Apply a bilateral filter to smooth the image and further reduce noise
filtered = cv2.bilateralFilter(closing, 9, 75, 75)

# Add text to the image
cv2.putText(filtered, 'PF1', (210, 120), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)

# Save the output image
img_copy = cv2.imwrite('Enhanced.jpg', filtered)

# Show the original and brightened image side by side
cv2.imshow('Original Image', img_gray)
cv2.imshow('Enhanced Image', img_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.subplot(121)
plt.imshow(img_gray)
plt.subplot(122)
plt.imshow(img_gray)
    
```
